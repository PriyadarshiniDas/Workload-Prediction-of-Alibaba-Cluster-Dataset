```python
df2=pd.read_csv('final_input_file.csv')
#df2
```

# Visualizing dataset

```python
fig=plt.figure(figsize=(12,8))
plt.title("Task vs Time")
plt.xlabel("Time interval")
plt.ylabel("Number of instances")
plt.plot(df2.time_interval, df2.no_of_instances)
plt.show()
```

# Exponential Smoothing

```python
#Computational Time
import time
time_start = time.clock()

#Code
z=0.5
Ft_prev=df2.no_of_instances[0]
print('Ft previous value=',Ft_prev)
X = df2.no_of_instances
predictions=list()
times=list()
SE=0
sumAPE=0
#sumMASE=0

for i in range (1,len(X)):
    Ft=np.int(z*df2.no_of_instances[i-1]+(1-z)*Ft_prev)
    predictions.append(Ft)
    Ft_prev=Ft
    times.append(df2.time_interval[i])
    print('Time, predicted, expected \n',df2.time_interval[i], Ft, df2.no_of_instances[i])
    e=abs(Ft-df2.no_of_instances[i])
    SE=SE+e*e
    APE=abs((Ft-df2.no_of_instances[i])/df2.no_of_instances[i])
    sumAPE=sumAPE+APE
    #ASE=abs((yhat-obs)/(test[t]-df2.no_of_instances[t-1]))
    #sumMASE=sumMASE+ASE
RMSE=math.sqrt(SE/len(X))
MAPE=(sumAPE/(len(X)))*100
#MASE=(sumMASE/(len(X)))
print('RMSE Value= ', RMSE)
print('MAPE Value= ', MAPE)
#print('MASE Value= ', MASE)
    
#print('Time= \n', times)
#print('Predictions= \n', predictions)

#Computational Time
time_elapsed = (time.clock() - time_start)
print('Computational Time=', time_elapsed)

# plot
fig=plt.figure(figsize=(12,8))
plt.title("Predicted vs Expected- - Exponential Smoothing")
plt.xlabel("Time interval")
plt.ylabel("Number of instances")
plt.plot(times, predictions, color='red', label='Predicted', linestyle='dashed')
plt.plot(df2.time_interval, df2.no_of_instances, color='blue', label='Original', linestyle='solid')
plt.legend()
plt.show()

    
```

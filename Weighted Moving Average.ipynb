```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.figsize']=10,6
import csv
import numpy as np
import math
```

```python
df2=pd.read_csv('final_input_file.csv')
#df2
```

# Visualizing Dataset

```python
fig=plt.figure(figsize=(12,8))
plt.title("Task vs Time")
plt.xlabel("Time interval")
plt.ylabel("Number of instances")
plt.plot(df2.time_interval, df2.no_of_instances)
plt.show()
```

```python
#Computational Time
import time
time_start = time.clock()

#Code
a1=0.1
a2=0.1
a3=0.1
a4=0.1
a5=0.6

X = df2.no_of_instances
Begin=5
predictions=list()
times=list()
SE=0
sumAPE=0
#sumMASE=0
#test=[X[i] for i in range(Begin,len(X)]
for i in range (Begin,len(X)):
    Ft=np.int(a1*df2.no_of_instances[i-5]+a2*df2.no_of_instances[i-4]+a3*df2.no_of_instances[i-3]+a4*df2.no_of_instances[i-2]+a5*df2.no_of_instances[i-1])
    predictions.append(Ft)
    Ft_prev=Ft
    times.append(df2.time_interval[i])
    print('Time, predicted, expected \n',df2.time_interval[i], Ft, df2.no_of_instances[i])
    e=abs(Ft-df2.no_of_instances[i])
    SE=SE+e*e
    APE=abs((Ft-df2.no_of_instances[i])/df2.no_of_instances[i])
    sumAPE=sumAPE+APE
    #ASE=abs((yhat-obs)/(test[t]-df2.no_of_instances[t-1]))
    #sumMASE=sumMASE+ASE
RMSE=math.sqrt(SE/len(X)-Begin)
MAPE=(sumAPE/(len(X)-Begin))*100
#MASE=(sumMASE/(len(X)))
print('RMSE Value= ', RMSE)
print('MAPE Value= ', MAPE)
#print('MASE Value= ', MASE)
    
#print('Time= \n', times)
#print('Predictions= \n', predictions)

#Computational Time
time_elapsed = (time.clock() - time_start)
print('Computational Time=', time_elapsed)

# plot
fig=plt.figure(figsize=(12,8))
plt.title("Predicted vs Expected- - Weighted Moving Average")
plt.xlabel("Time interval")
plt.ylabel("Number of instances")
plt.plot(times, predictions, color='red', label='Predicted', linestyle='dashed')
plt.plot(df2.time_interval, df2.no_of_instances, color='blue', label='Original', linestyle='solid')
plt.legend()
plt.show()

    
```
